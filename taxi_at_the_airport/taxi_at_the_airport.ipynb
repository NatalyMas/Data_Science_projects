#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Построим модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Основные шаги:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


## Подготовка

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from sklearn.model_selection import train_test_split
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_score, RandomizedSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from sklearn.metrics import make_scorer

from sklearn.dummy import DummyRegressor
from statsmodels.tsa.stattools import adfuller

import warnings
warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (15,9)
```

```python
pth1 = '/datasets/taxi.csv'
pth2 = 'taxi.csv'

if os.path.exists(pth1):
    data = pd.read_csv(pth1, index_col=[0], parse_dates=[0])
elif os.path.exists(pth2):
    query_1 = pd.read_csv(pth2)
else:
    data = pd.read_csv(pth2, index_col=[0], parse_dates=[0])
```

Данные загружены. Датасет представляет собой временной ряд количества заказов, временной интервал - с марта по сентябрь 2018г, даты и время расположены в хронологическом ли порядке.

В общем, заказов становится больше. Возможно, заказов становится больше из-за смены времен года, например, летом больше людей летают в отпуск (если предположить, что данные по северному полушарию), и эта ситуация с летним подъемом повторяется каждый год. Но увидеть мы этого не можем, т.к. данные есть только за полгода.

Каждый день картина примерно одинаковая. На суточном графике видно минимум у 6 часов утра и максимум в полночь. Возможно, это связано с тем, что интервал 4-7 утра очень неудобен в смысле времени прибытия/вылета рейсов, из-за чего людей на таких рейсах меньше. А поздно вечером удобно и для вылета (можно надеяться поспать в самолете), и для прибытия (на такси домой и спать).

Создадим признаки: календарные, отстающие значения и скользящее среднее. Затем разделим датасет на обучающую и тестовую выборки и отделим целевой признак. Количество отстающих значений возьмем 23, т.к. каждые 24 часа ситуация примерно повторяется, и условно 1-й и 25-й лаг будут сильно коррелировать.

```python
def make_features(data, max_lag, rolling_mean_size):
    
    data_func = data.copy()

    data_func['day'] = data.index.day
    data_func['dayofweek'] = data.index.dayofweek
    data_func['hour'] = data.index.hour
    data_func['deep_sleep_time'] = data_func['hour'].isin([5,6,7])*1
    data_func['party_time'] = data_func['hour'].isin([23,0,1])*1
    
    for lag in range(1, max_lag + 1):
        data_func['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data_func['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    return data_func
```

```python
data_f = make_features(data, 23, 4)

train, test = train_test_split(data_f, shuffle=False, test_size=0.1)
train = train.dropna()

y_train = train['num_orders']
train = train.drop('num_orders', axis=1)
y_test = test['num_orders']
test = test.drop('num_orders', axis=1)
train.head()
```

Данные готовы к обучению.

## Обучение

Создадим списки для таблицы результатов и инструмент для корректного разделения фолдов по времени.

```python
model_list = ['LR', 'Ridge', 'RFR', 'LGB', 'CBR']
result_train_list = []
result_test_list = []
fold_splitter = TimeSeriesSplit(n_splits=3)
```

```python
#модели
model_LR = LinearRegression()
model_Ridge = Ridge(random_state=12345)
model_RFR = RandomForestRegressor(random_state=12345)
model_LGB = LGBMRegressor(random_state=12345)
model_CBR = CatBoostRegressor(n_estimators=500, random_state=12345)
```

```python
#наборы параметров
param_Ridge = {'alpha': [0.1, 1, 10, 100, 1000]}
param_RFR = {'n_estimators': [20, 100, 400, 1000, 2000, 3000], 'max_depth': [7, 11, 19, 29, 35], 
             'min_samples_split': [2, 3, 5, 8]}
param_LGB = {'n_estimators': [50, 100, 400, 1000], 'max_depth': [1, 2, 5, 7]}
param_CBR = {'depth': [2, 3, 7], 'learning_rate': [0.001, 0.03, 0.05, 0.08, 0.1, 0.3]}
```

Попробуем сначала линейную регрессию.

```python
rmse_LR = cross_val_score(model_LR, train, y_train, cv=fold_splitter, scoring='neg_root_mean_squared_error', n_jobs=-1)
result_train_list.append(-1*rmse_LR.mean())

print(-1*rmse_LR.mean())
```

Попробуем другие модели - Ridge, случайный лес, LGBM и CatBoost с подбором гиперпараметров.

```python
#функция для работы модели с подбором параметров
def model_run(model, params, x_train, y_train):
    model_opt = RandomizedSearchCV(model, params, random_state=12345, cv=fold_splitter, n_jobs=-1, 
                                   scoring='neg_root_mean_squared_error', verbose=10)
    model_opt.fit(x_train, y_train)
    return -1*model_opt.best_score_, model_opt.best_estimator_    
```

```python
result_Ridge = model_run(model_Ridge, param_Ridge, train, y_train)
result_train_list.append(result_Ridge[0])
print(result_Ridge)
```

```python
result_RFR = model_run(model_RFR, param_RFR, train, y_train)
result_train_list.append(result_RFR[0])
print(result_RFR)
```

```python
result_LGB = model_run(model_LGB, param_LGB, train, y_train)
result_train_list.append(result_LGB[0])
print(result_LGB)
```

```python
result_CBR = model_run(model_CBR, param_CBR, train, y_train)
result_train_list.append(result_CBR[0])
print(result_CBR)
```

```python
result_tab_train = pd.DataFrame({'RMSE train': result_train_list}, index=model_list)
result_tab_train
```

Хуже всего себя показали линейные модели. Лучший результат у CatBoost. Проверим его на тестовой выборке.

## Тестирование

Посмотрим, какой результат показывает на тестовой выборке лучшая модель.

```python
pred = result_CBR[1].predict(test)
rmse = mean_squared_error(pred, y_test, squared=False)
print(rmse)
```

Построим в одних осях предсказания модели и реальные значения.

```python
plt.plot(pred, label="Предсказания")
plt.plot(y_test.values, label="Реальные значения")
plt.legend()
plt.title('Результаты')
plt.grid(True)
```

Видно, что модель плохо справляется с пиковыми значениями.

Проверим модель на адекватность с помощью DummyRegressor.

```python
model_Dummy =  DummyRegressor()
model_Dummy.fit(train, y_train)
pred_Dummy = model_Dummy.predict(test)
rmse_Dummy = mean_squared_error(pred_Dummy, y_test, squared=False)
rmse_Dummy
```

Метрика у константной модели в 2 раза больше, проверка пройдена.

Построим таблицу важности признаков итоговой модели.

```python
feature_imp = pd.Series(result_CBR[1].feature_importances_, train.columns)
fig, ax = plt.subplots(figsize=(16,14))
feature_imp.plot.bar(ax=ax)
ax.set_title('Важность признаков для модели CatBoostRegressor')
ax.set_ylabel('Важность')
ax.set_xlabel('Признаки')
fig.tight_layout()
```
